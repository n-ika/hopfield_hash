```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_path = '/Users/nika/Desktop/ae_schapiro/stats/'
files = os.listdir(root_path)

results = {"epochs":[],"a2a" : [],"b2b" : [],"a2b" : [],"b2a" : []}

for f in files:
    df = pd.read_csv(root_path+f)
    results['a2a'].append(df.a2a.mean())
    results['b2b'].append(df.b2b.mean())
    results['b2a'].append(df.b2a.mean())
    results['a2b'].append(df.a2b.mean())
    results['epochs'].append(int(f.split('_')[1].split('.csv')[0]))
```

```python
stats = pd.DataFrame(results)
stats.sort_values('epochs', inplace=True)
stats
```

```python
plt.plot(stats['epochs'],stats['a2a'],label='a2a')
plt.plot(stats['epochs'],stats['b2b'],label='b2b')
plt.plot(stats['epochs'],stats['a2b'],label='a2b')
plt.plot(stats['epochs'],stats['b2a'],label='b2a')
plt.legend()
plt.savefig('./schapiro_ae_reproduction.jpg')
```

```python
res = {"epochs":[],"a2a" : [],"b2b" : [],"a2b" : [],"b2a" : []}

for f in files:
    df = pd.read_csv(root_path+f)
    df.a2a[df.a2a < 0.5] = 0
    df.a2a[df.a2a > 0.5] = 1
    df.b2b[df.b2b < 0.5] = 0
    df.b2b[df.b2b > 0.5] = 1
    df.a2b[df.a2b < 0.5] = 0
    df.a2b[df.a2b > 0.5] = 1
    df.b2a[df.b2a < 0.5] = 0
    df.b2a[df.b2a > 0.5] = 1
    res['a2a'].append(df.a2a.mean())
    res['b2b'].append(df.b2b.mean())
    res['b2a'].append(df.b2a.mean())
    res['a2b'].append(df.a2b.mean())
    res['epochs'].append(int(f.split('_')[1].split('.csv')[0]))
```

```python
res = pd.DataFrame(res)
res.sort_values('epochs', inplace=True)
```

```python
res_f = res[res.epochs<=11]
```

```python
plt.plot(res_f['epochs'],res_f['a2a'],label='a2a')
plt.plot(res_f['epochs'],res_f['b2b'],label='b2b')
plt.plot(res_f['epochs'],res_f['a2b'],label='a2b')
plt.plot(res_f['epochs'],res_f['b2a'],label='b2a')
plt.legend()
plt.show()
```
